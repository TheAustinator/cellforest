### Testing performance

```python
# checking that the result is the same
print(np.mean(rna._matrix.toarray(), axis=0))
print(np.mean(rna._matrix, axis=0))
print(rna._matrix.mean(axis=0))
```

```python
print(rna._matrix.tocsc().mean(axis=0))  # CSC format for fast arithmetics
```

```python
import time

NUM_EXP = 100000
```

```python
start = time.time()
for _ in range(NUM_EXP):
    np.mean(rna._matrix.toarray(), axis=0)

stop = time.time()
print(stop - start)
```

```python
start = time.time()
for _ in range(NUM_EXP):
    np.mean(rna._matrix, axis=0)

stop = time.time()
print(stop - start)
```

```python
start = time.time()
for _ in range(NUM_EXP):
    rna._matrix.mean(axis=0)

stop = time.time()
print(stop - start)
```

```python
start = time.time()
for _ in range(NUM_EXP):
    rna._matrix.tocsc().mean(axis=0)

stop = time.time()
print(stop - start)
```
